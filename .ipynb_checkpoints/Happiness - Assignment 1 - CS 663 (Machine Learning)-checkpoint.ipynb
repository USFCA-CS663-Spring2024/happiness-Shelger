# Question 2:
1. What is the happiest country in 2022? In 2013? Of all time? What is the least happy country in
2022, 2013 and of all time?
    Happiest Country in 2022: Finland
    Least Happy Country in 2022: Afghanistan
    Happiest Country in 2013: Canada
    Least Happy Country in 2013: Syria
    Happiest Country all time: Denmark
    Least Happy Country all time: Afghanistan
    I calculated the even of each countries to get the least and happiest country. I think there can be other criteria such as median.

```python
def happiest_and_least_happy_year(year):
    df_year = df[df['year'] == year]
    happiest_country = df_year.loc[df_year['Life Ladder'].idxmax()]['Country name']
    least_happy_country = df_year.loc[df_year['Life Ladder'].idxmin()]['Country name']
    return happiest_country, least_happy_country

happiest_2022, least_happy_2022 = happiest_and_least_happy_year(2022)
happiest_2013, least_happy_2013 = happiest_and_least_happy_year(2013)
```

```python
df_2023 = df[df['year'] == 2023]
print(df_2023)
```

```python
print(f"Happiest Country in 2022: {happiest_2022}")
print(f"Least Happy Country in 2022: {least_happy_2022}")
print(f"Happiest Country in 2013: {happiest_2013}")
print(f"Least Happy Country in 2013: {least_happy_2013}")
```

```python
average_scores = df.groupby('Country name')['Life Ladder'].mean()
happiest_all_time = average_scores.idxmax()
least_happy_all_time = average_scores.idxmin()
```

```python
print(f"Happiest Country: {happiest_all_time}")
print(f"Least Happy Country: {least_happy_all_time}")
```

# Question 3:
1. Of the countries in this dataset, which country had the largest increase in happiness from its start
of participation to 2022? Which had the largest decrease in happiness?
    Country with the largest increase in happiness: Congo (Brazzaville)
    Country with the largest decrease in happiness: Lebanon

```python
df_2022 = df[df['year'] == 2022]
first_year_df = df.groupby('Country name').first().reset_index()
merged_df = pd.merge(first_year_df[['Country name', 'Life Ladder']], df_2022[['Country name', 'Life Ladder']], on='Country name', suffixes=('_first', '_2022'))
merged_df['change_in_happiness'] = merged_df['Life Ladder_2022'] - merged_df['Life Ladder_first']
```

```python
print(merged_df)
```

```python
largest_increase_country = merged_df.loc[merged_df['change_in_happiness'].idxmax()]['Country name']
largest_decrease_country = merged_df.loc[merged_df['change_in_happiness'].idxmin()]['Country name']
```

```python
print(f"Country with the largest increase in happiness: {largest_increase_country}")
print(f"Country with the largest decrease in happiness: {largest_decrease_country}")
```
